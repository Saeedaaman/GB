```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name = "gpt2"  # You can choose other GPT-2 variants depending on your needs
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
model_name = "gpt2"  # You can choose other GPT-2 variants depending on your needs
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    input_ids = tokenizer.encode("You: " + user_input, return_tensors="pt")
    response_ids = model.generate(input_ids, max_length=100, num_return_sequences=1, pad_token_id=50256)

    chatbot_response = tokenizer.decode(response_ids[0], skip_special_tokens=True)
    print("Chatbot:", chatbot_response)
```
