```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
india_runs = np.random.randint(50, 150, size=11)
pakistan_runs = np.random.randint(50, 150, size=10)

# Create a DataFrame for the data
data = pd.DataFrame({
    'Team': ['India'] * len(india_runs) + ['Pakistan'] * len(pakistan_runs),
    'Runs': np.concatenate([india_runs, pakistan_runs])
})

# Save data to a CSV file
data.to_csv('cricket_match_data.csv', index=False)

# Read the data from the CSV file
data = pd.read_csv('cricket_match_data.csv')

# Extract runs scored by players from both teams
india_runs = data[data['Team'] == 'India']['Runs']
pakistan_runs = data[data['Team'] == 'Pakistan']['Runs']

# Create a histogram
plt.hist([india_runs, pakistan_runs], bins=10, alpha=0.7, color=['blue', 'green'], label=['India', 'Pakistan'])

# Adding labels and title
plt.xlabel('Runs Scored')
plt.ylabel('Number of Players')
plt.title('Runs Scored by Players - India vs Pakistan')

# Add a legend
plt.legend()

# Add text annotation for Pakistan's victory
plt.annotate('Pakistan won', xy=(110, 4), xytext=(120, 6),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

# Display the histogram
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
sectors = ['Education', 'Healthcare', 'Infrastructure', 'Agriculture', 'Tourism']
budget_allocation = np.array([2500000, 1800000, 1500000, 1200000, 900000])  # Budget allocation for each sector
expenditures = np.random.randint(500000, 1000000, size=len(sectors))  # Random expenditures for each sector

# Create a DataFrame for the data
data = pd.DataFrame({
    'Sector': sectors,
    'Budget Allocation': budget_allocation,
    'Expenditure': expenditures
})

# Calculate total expenditure and budget allocation
total_expenditure = data['Expenditure'].sum()
total_budget_allocation = data['Budget Allocation'].sum()

# Calculate the percentage of budget allocation used for each sector
data['Percentage Used'] = (data['Expenditure'] / data['Budget Allocation']) * 100

# Save data to a CSV file
data.to_csv('gilgit_baltistan_budget.csv', index=False)

# Read the data from the CSV file
data = pd.read_csv('gilgit_baltistan_budget.csv')

# Display total expenses
print(f"Total Expenditure of Gilgit-Baltistan: {total_expenditure:.2f}")
print(f"Total Budget Allocation: {total_budget_allocation:.2f}\n")

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(data['Expenditure'], labels=data['Sector'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title("Expenditure Distribution by Sector - Gilgit-Baltistan")
plt.legend(loc='best')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.

# Display the pie chart
plt.show()
```
